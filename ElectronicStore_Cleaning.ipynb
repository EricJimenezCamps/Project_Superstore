```python
df_store.isnull().sum()
```

```python
df_store[df_store['category_id'].isnull()]
```

```python
#Son muchos datos nulos... que podemos hacer con ellos?? Buscaremos no eliminarlos.
#Substituiremos los nulos a 'sin_datos_categoria_ID'
```

```python
df_store['category_id'].fillna('sin_categoria_id', inplace=True)
```

```python
df_store[df_store['category_code'].isnull()]
```

```python
#Son muchos datos nulos... que hacemos con ellos?
#No podemos saber cual seria el codigo de categoria... Lo podriamos saber con el product_id o category_id?
#Asignaremos a los valores nulos 'sin_categoria_code'
```

```python
df_store['category_code'].fillna('sin_categoria_code', inplace=True)
```

```python
df_store[df_store['brand'].isnull()]
```

```python
#Nos encontramos con el mismo problema que en los casos anteriores. 
#Substituiremos los valores nulos por 'sin_marca'
```

```python
df_store['brand'].fillna('sin_marca', inplace=True)
```

```python
df_store[df_store['price'].isnull()]
```

```python
#En este caso, los valores nulos en precio, los podemos substituir por la media del 
#conjunto de dstos.
```

```python
mean_price= df_store['price'].mean()
round(mean_price,2)
```

```python
df_store['price'].fillna(mean_price, inplace=True)
```

```python
df_store[df_store['user_id'].isnull()]
```

```python
#Optamos por seguir el mismo planteamiento que con los otros nulos
#sobre los que no podemos saber su valor.
#Substituiremos los valores nulos por 'sin_usuario_ID'
```

```python
df_store['user_id'].fillna('sin_user_ID', inplace=True)
```

```python
#Revisamos que no tengamos ya ningún valor nulo.
df_store.isnull().sum()
```
